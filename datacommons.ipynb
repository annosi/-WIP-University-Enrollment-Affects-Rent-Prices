```python
df_county = dc.build_multivariate_dataframe(counties, ['Count_Person', 'Count_HousingUnit_OwnerOccupied','Count_HousingUnit_RenterOccupied'])
```

```python
df_county.head()
```

```python
def add_name_col(df):
  # Add a new column called name, where each value is the name for the place dcid in the index.
  df['name'] = df.index.map(dc.get_property_values(df.index, 'name'))
  
  # Keep just the first name, instead of a list of all names.
  df['name'] = df['name'].str[0]
```

```python
add_name_col(df_county)
df_county.head()
```

```python
df_time = dc.build_time_series_dataframe(counties, 'Count_Person')
```

```python
add_name_col(df_time)
df_time.head()
```

```python
df_owner = dc.build_time_series_dataframe(counties, 'Count_HousingUnit_OwnerOccupied')
```

```python
add_name_col(df_owner)
df_owner.head()
```

```python
df_renter = dc.build_time_series_dataframe(counties, 'Count_HousingUnit_RenterOccupied')
df_renter.head()
```

```python
add_name_col(df_renter)
df_renter.head()
```
